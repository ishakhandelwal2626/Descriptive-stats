7. Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).
Company A: variance = 6
Company B: variance = 15
 Which company is more consistent, and why?
- Variance measures how spread out the data is from the average.

  Lower variance - less variation - more consistency

  Higher variance - more fluctuation - less consistency

  Company A: Variance is 6
  Means it is more consitent. Because,
  Company A’s delivery delays are closer to the average delay.

  Company B shows greater variability, meaning delivery times fluctuate more.

  A variance of 15 indicates delays are less predictable than a variance of 6.
